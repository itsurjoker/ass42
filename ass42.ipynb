```python
#Q4.
#To calculate the Pearson correlation coefficient between two variables, you need to follow these steps:

#    Calculate the mean (average) of both variables.
#    Calculate the difference between each individual data point and the mean for both variables.
#    Multiply the differences for each individual and sum them up.
#    Calculate the squared differences for each individual for both variables and sum them up.
#    Divide the sum of the multiplied differences (step 3) by the square root of the product of the sum of squared differences for both variables (step 4).

#Let's assume that the researcher has already collected the data on the number of hours individuals spend watching television per day and their level of physical activity, and the data is represented by two arrays: hours_of_tv and physical_activity.

import numpy as np

# Assuming the data is already collected and stored in arrays
hours_of_tv = [4, 3, 2, 5, 6, ...]  # List of hours of TV per day
physical_activity = [120, 140, 90, 160, 110, ...]  # List of physical activity levels

# Calculate the mean of both variables
mean_hours = np.mean(hours_of_tv)
mean_activity = np.mean(physical_activity)

# Calculate the differences between each data point and the mean for both variables
diff_hours = hours_of_tv - mean_hours
diff_activity = physical_activity - mean_activity

# Calculate the sum of the product of the differences
sum_product_diff = np.sum(diff_hours * diff_activity)

# Calculate the sum of squared differences for both variables
sum_squared_diff_hours = np.sum(diff_hours ** 2)
sum_squared_diff_activity = np.sum(diff_activity ** 2)

# Calculate the Pearson correlation coefficient
pearson_corr = sum_product_diff / np.sqrt(sum_squared_diff_hours * sum_squared_diff_activity)

print("Pearson correlation coefficient:", pearson_corr)
```
